```python
node_ids = {'node':[]}

for node in df1['Source']:
  if node not in node_ids['node']:
    node_ids['node'].append(node)

for node in df1['Target']:
  if node not in node_ids['node']:
    node_ids['node'].append(node)
```

```python
node_ids = pd.DataFrame(node_ids)
```

```python
ids = np.arange(0, node_ids.shape[0])
```

```python
node_ids['ids'] = ids
```
